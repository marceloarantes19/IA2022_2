```python
# Sexo - Feminino (1) e Masculino (2)
# País - Alemanha (1), França (2) e Inglaterra (3)
# Idade - Mantém
# Comprar - Sim (1), Não (2)

from sklearn import tree
features = [
				[1,2,25],
				[1,3,21],
				[1,2,23],
				[1,3,34],
				[1,2,30],
				[2,1,21],
				[2,1,20],
				[1,1,18],
				[1,2,34],
				[2,2,55]
            ]
labels = [1,1,1,1,2,2,2,2,2,2]
classif = tree.DecisionTreeClassifier()
classif.fit(features, labels)
tree.plot_tree(classif)
#print("2 - ", classif.predict([[2,2,55]]))
#for i in range(1,60):
#	print(i, " - ", classif.predict([[2,2,i]])
```
